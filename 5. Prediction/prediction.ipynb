```python
# ============================================================
# ⚙️ Configuration & Imports
# ============================================================
import os
import pandas as pd
import joblib

# -------------------------
# Configurable paths
# -------------------------

models_folder = "../4. Models"
input_file = "../0. Datasets/patients_to_predict.csv"   # <-- Change this to your actual input file path
output_file = "diabetes_prediction.csv"
```

```python
# ============================================================
# 🧠 Load Trained Model
# ============================================================
# List available models
available_models = [m for m in os.listdir(models_folder) if m.endswith(".pkl")]
print("Available models:", available_models)

# Choose a model (you can change the index or specify directly)
model_name = available_models[0]
model_path = os.path.join(models_folder, model_name)

# Load the model
model = joblib.load(model_path)
print(f"✅ Loaded model: {model_name}")
```

```python
# ============================================================
# 📊 Load Input Data
# ============================================================
# The dataset should have the same columns used for training, except 'has_diabetes'
df = pd.read_csv(input_file)

print("📋 Input data preview:")
display(df.head())
```

```python
# ============================================================
# 🔮 Make Predictions
# ============================================================
# Generate predictions (0 = No Diabetes, 1 = Has Diabetes)
predictions = model.predict(df)

# Create a DataFrame for output
output_df = pd.DataFrame({
    "patient_id": [str(i + 1).zfill(5) for i in range(len(predictions))],
    "has_diabetes": predictions.astype(int)
})

print("✅ Prediction complete:")
display(output_df.head())
```

```python
# ============================================================
# 💾 Save Predictions to CSV
# ============================================================
output_df.to_csv(output_file, index=False)
print(f"✅ Predictions saved successfully to '{output_file}'")
```

# PREDICT 1 PATIENT (Manual Input)

---------------------------------------------------------------
📘 Column Format (Input)
---------------------------------------------------------------
patient_id:         int  
has_diabetes:       boolean  
Age:                int  
Gender:             ('Male', 'Female')  
Hypertension:       boolean  
Heart Disease:      boolean  
Smoking History:    ('Non-Smoker', 'Past', 'Current', None)  
BMI:                float  
HbA1c:              float  
Random Glucose:     float  

---------------------------------------------------------------
🔢 Conversions (Mapping)
---------------------------------------------------------------
Gender:
    Male   → 0
    Female → 1

Hypertension:
    False → 0
    True  → 1

Heart Disease:
    False → 0
    True  → 1

Smoking History:
    Non-Smoker → 0
    Past       → 1
    Current    → 2
    None/NaN   → -1

has_diabetes:
    False → 0
    True  → 1

```python
# ============================================================
# 👤 Predict for a Single Patient (Manual Input)
# ============================================================
# Example: Define a single patient's data as a dictionary
# ⚠️ Make sure these feature names match your training data exactly!
single_patient = {
    #"patient_id": 00000, 
    #"has_diabetes": 0, 
    "Age": 20,
    "Gender": 1,
    "Hypertension": 0, 
    "Heart Disease": 1, 
    "Smoking History": 0,
    "BMI": 25,
    "HbA1c": 4.5,
    "Random Glucose": 170
}



# Convert to DataFrame
patient_df = pd.DataFrame([single_patient])

# Predict using the loaded model
single_pred = model.predict(patient_df)[0]

print("============================================")
print("🧍 Single Patient Prediction Result")
print("============================================")
print(f"Input Data: {single_patient}")
print(f"Prediction (has_diabetes): {int(single_pred)}")
print("============================================")
```
